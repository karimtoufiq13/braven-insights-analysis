```python
# Load the provided dataset for analysis
file_path = r"C:\Users\ktouf\Downloads\Data Analyst Activity Data Set.csv"

# Load the dataset into a Pandas DataFrame
data = pd.read_csv(file_path)

# Display the first few rows and basic information about the dataset
data_head = data.head()
data_info = data.info()

# Check for unique column names
column_names = data.columns.tolist()

data_head, column_names
```

```python
# We will evaluate the satification report and get some greater details about the results 
satisfaction_counts = data[
    "How much do you think Braven has improved your chances of getting the job or graduate school admission you want after college?"
].value_counts()

print("\nSatisfaction Breakdown:")
print(satisfaction_counts)

# Satisfaction by Program (Site)
if "Program" in data.columns:
    satisfaction_by_site = data.groupby("Program")[
        "How much do you think Braven has improved your chances of getting the job or graduate school admission you want after college?"
    ].value_counts(normalize=True)
    print("\nSatisfaction by Program:")
    print(satisfaction_by_site)
```

Majority Positive Impact:
76% of participants felt that Braven improved their chances either "quite a bit" or "a great deal" (107 + 101 = 208 out of 330 total participants).
This highlights that the program is generally well-received and positively impacts participants' career prospects.
Moderate to Little Impact:
23% of participants reported only a "moderate" improvement or less.
While these numbers are lower, they indicate room for improvement to ensure a more consistently positive experience.
Low Satisfaction:
5% (17 participants) reported that the program did not improve their chances at all.
Understanding why these participants felt this way could provide actionable insights to improve the program.


```python
# Lets now take a look at NPS Score Analysis
nps_counts = data["NPS Score"].value_counts()
print("\nNPS Score Distribution:")
print(nps_counts)

# Calculate NPS Percentages
nps_promoters = len(data[data["NPS Score"] >= 9]) / len(data) * 100
nps_passives = len(data[(data["NPS Score"] >= 7) & (data["NPS Score"] <= 8)]) / len(data) * 100
nps_detractors = len(data[data["NPS Score"] <= 6]) / len(data) * 100

print("\nNPS Categories:")
print(f"Promoters: {nps_promoters:.2f}%")
print(f"Passives: {nps_passives:.2f}%")
print(f"Detractors: {nps_detractors:.2f}%")
```

Key Observations
High Proportion of Detractors:

The percentage of Detractors (43.64%) is higher than both Promoters (30.30%) and Passives (26.06%).
This indicates a potential issue with certain aspects of the program, as a significant portion of participants are either dissatisfied or indifferent.
Promoters Outnumber Passives:

Although Promoters (30.30%) outnumber Passives (26.06%), the gap is not large, suggesting room to convert Passives into Promoters.
Balanced Spread Across Scores:

The distribution shows a balanced spread across the full range of scores (0–10), which suggests diverse participant experiences. However, the high number of 5s and 6s indicates a middle-ground dissatisfaction that should be explored.

```python
import seaborn as sns

# Filter for participants with career-aligned roles
career_aligned_data = data[
    data["Was your role related to your career interests?"] == "Yes"
]

# Group by NPS categories and calculate mean networking scores
nps_networking = data.groupby(
    pd.cut(data["NPS Score"], bins=[0, 6, 8, 10], labels=["Detractors", "Passives", "Promoters"])
)["Networking Self Efficacy"].mean()

# Scatterplot: Career-aligned role satisfaction
plt.figure(figsize=(8, 6))
sns.scatterplot(
    x=career_aligned_data["NPS Score"],
    y=career_aligned_data["Networking Self Efficacy"],
    color="#0033A0",
    edgecolor="black"
)
plt.title("NPS Score vs. Networking Self-Efficacy (Career-Aligned Roles)", fontsize=16, color="#696969")
plt.xlabel("NPS Score", fontsize=12)
plt.ylabel("Networking Self-Efficacy", fontsize=12)
plt.grid(visible=True, linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()

# Bar chart: Networking self-efficacy by NPS categories
nps_networking.plot(
    kind="bar",
    color=["#8B0000", "#FFD700", "#0033A0"],
    edgecolor="black",
    figsize=(8, 6)
)
plt.title("Average Networking Self-Efficacy by NPS Category", fontsize=16, color="#696969")
plt.xlabel("NPS Category", fontsize=12)
plt.ylabel("Average Networking Score", fontsize=12)
plt.xticks(rotation=0, fontsize=10)
plt.tight_layout()
plt.show()

nps_networking
```
